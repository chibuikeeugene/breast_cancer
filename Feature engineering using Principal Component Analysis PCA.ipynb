```python
#import the necessary library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read the dataset using pandas read csv method
breast_cancer = pd.read_csv('breast-cancer-data.csv')
#get the first few observations
breast_cancer.head()
```

```python
#call pandas display option method to get all data columns
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
# get the info from the data
breast_cancer.info()
```

```python
#obtain the descriptive statistics of the data
breast_cancer.describe()
```

```python
#extract target variable
Target_data = breast_cancer['diagnosis']
```

```python
#verify changes
Target_data.head()
```

```python
#convert target variable to ones and zeros using label encoder for better analysis
#import label encoder from sklearn preprocessing package
from sklearn.preprocessing import LabelEncoder
#create an object instance of the class
encoder = LabelEncoder()
Target_data = encoder.fit_transform(Target_data)
#verify changes
Target_data
```

```python
#drop the diagnosis column 
features = breast_cancer.drop(['diagnosis'], axis = 1)
features.head()
```

# applying principal component analysis for feature extraction

Let's use PCA to find the first two principal components, and visualize the data in this new, two-dimensional space, with a single scatter-plot

Scale data so that each feature has a single unit variance.

```python
#import standard scalar from sklearn preprocessing library
from sklearn.preprocessing import StandardScaler
#create an object instance
scalar = StandardScaler()
#fit_transform the data
scaled_data = scalar.fit_transform(features)
```

```python
#Now obtaining the first two principal components of features
from sklearn.decomposition import PCA
#instantiate the object of the class
pca = PCA(n_components= 2)
#fit transform the scaled data
x_pca = pca.fit_transform(scaled_data)
```

```python
#verify shape of original data
scaled_data.shape
```

```python
#verify shape of final data
x_pca.shape
```

```python
#Plot Scatter plot to visualize the two components
plt.figure(figsize = (9,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = Target_data, cmap = 'viridis' )
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

NB: The components correspond to combinations of the original features. The components themselves are stored as an attribute of the fitted PCA object:

```python
#obtaining the explained variance of the components
pca.explained_variance_ratio_
```

```python
#lets alter our number of components to three and see if there's any differnce
pca_3 = PCA(n_components=3)
x_pca3 = pca_3.fit_transform(scaled_data)
x_pca3.shape
```

Checking the accuracy of 2 vs 3 components

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_ouptut, test_output = train_test_split(x_pca, Target_data, test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LogisticRegression
logisticsreg = LogisticRegression()
logisticsreg.fit(train_data, train_ouptut)
```

```python
#obtain accuracy score
logisticsreg.score(test_data, test_output)
```

```python
#obtaining that of the three components
train_data, test_data, train_ouptut, test_output = train_test_split(x_pca3, Target_data, test_size = 0.3, random_state = 100)
logisticsreg.fit(train_data, train_ouptut)
logisticsreg.score(test_data, test_output)
```

Conclusion - three principal components produced a higher accuracy than two components
